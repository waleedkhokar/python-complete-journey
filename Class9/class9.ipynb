# while loop
* while logic: #true/false
    * loop_body

```python
flag : bool = True

current_number : int = 1
while flag:
    print(f"Curent number is : {current_number}")
    current_number += 1
    if current_number == 10:
        break
```

```python
l1 : list[list] = [100,200,300]

index: int = 0 

while index < len(l1):
    print(f"current_number is : {index} and list value and  {l1[index]}")
    index += 1
```

```python
data: list[dict[str,str]] = []
flag : bool = True

while flag:
    print("write a quite or stop to this paper")
    name: str = input("your good name ? \t:")
    education: str = input("your good name ? \t:")
    
    if name  in ['exit', 'stop', 'close'] or education  in ['exit', 'stop', 'close']:
        flag=False
        break
    data.append({"name" : name,
             "education": education})
            
display(data)    
```

```python
prompt : int= "\nTell me something, and I will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program. "
active = True
while active:
        message = input(prompt)
if message == 'quit':
        active = False
else:
     print(message)
```

```python
prompt = "\nPlease enter the name of a city you have visited:"
prompt += "\n(Enter 'quit' when you are finished.) "
while True:
     city = input(prompt)
     if city == 'quit':
          break
     else:
          print(f"I'd love to go to {city.title()}!")
```
