```python
from  typing import Dict, Union, Optional

import pprint

key = Union[int,str]
Value = Union[int, str, list, tuple, set]

# import pprint # for  beautiful printout 
# display is used mostly for jupiter notebook google 
# list  
data: Dict[key, Value] = {
    "fname": "Muhhmad", 
                        "name": "eale", 
                        "education": "BSCS",
                        "abc": [1,2,5],
                        "cde": {"a":1, "b": 2}
                        }

# through lopp
for k,v in data.items():
    print(k,v)
```
